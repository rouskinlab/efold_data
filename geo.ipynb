```python
files = []
for f in os.walk('/Users/yvesmartin/data/geo/pri-miRNA'):
    files.extend(f[2])

files_pri = sorted([f for f in files if f.endswith('R1_001.fastq.gz')])

files = []
for f in os.walk('/Users/yvesmartin/data/geo/mRNA'):
    files.extend(f[2])
    
files_mRNA = sorted([f for f in files if f.endswith('R1_001.fastq.gz')])

import numpy as np
idx = np.argsort(files_pri + files_mRNA)

fastas = np.array([None]*len(files_pri) + ['mRNA.fasta']*len(files_mRNA))[idx].tolist()

with open('fastas.tsv', 'w') as f:
    for fasta in fastas:
        if fasta is not None:
            f.write(fasta + '\n')
        else:
            f.write('\n')
```

```python
pri = ['1-5_1_A_S1',
 '1-5_1_B_S2',
 '1-5_2_A_S3',
 '1-5_2_B_S4',
 '1-5_3_A_S5',
 '1-5_3_B_S6']

class Title:
    def __init__(self, name):
        if '--' in name:
            name = name.split('--')[0]
        self.name = name
        self.concentration = self.process_concentration(name)
        self.experiment = 'pri-miRNA' if name[:10] in pri else 'mRNA'
        self.plate_number = name.split('_')[1]
        if self.concentration != 'Untreated':
            self.replicate = name.split('_')[2]
        else:
            self.replicate = None
        
    def process_concentration(self, name):
        conc = name.split('_')[0]
        if conc == 'Un':
            return 'Untreated'
        else:
            return '[DMS]:' + conc.replace('-','.') + '%'
        
    def __str__(self):
        args = {
            'experiment: ': self.experiment,
            'plate_number: ': self.plate_number,
            ' ': self.concentration,
            'replicate: ': self.replicate
        }
        if self.replicate is None:
            del args['replicate: ']
        out = ', '.join([f'{k}{v}' for k,v in args.items()])
        if '--' in self.name:
            out += 'TODO'
        return out

titles = open('r1.tsv').read().split('\n')
titles = [Title(t) for t in titles]
with open('titles.tsv', 'w') as f:
    for t in titles:
        f.write(str(t) + '\n')
        
        
```
